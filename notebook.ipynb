```python
def checknull(t):
    x=0
    for i in range(len(t)):
        if(t[i]==-200):
            x=x+1
        elif(t.isna()[i]==True):
            x=x+1
        else:
            pass
    return x  
```

```python
for i in df.columns:
    print(i,"->",checknull(df[i])) 
```

```python
row,col = df.shape
count_suc=0
for i in range (col):
    if(checknull(df[df.columns[i]])!=0):
        df[df.columns[i]]=df[df.columns[i]].replace(-200,np.nan)
        for j in range (row):
            if(df[df.columns[i]].isna()[j]==True):
                count_suc +=1
            else:
                pass
print(count_suc)
```

# Machine Learning

## Linear Regression

```python
X = df[features_df].drop('RH',axis=1)
```

```python
y = df['RH']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score,mean_absolute_error,mean_squared_error,r2_score
```

```python
X_std = StandardScaler().fit_transform(X)
```

```python
train_X, test_X, train_y,test_y = train_test_split(X_std,y,test_size=0.3,random_state=40)
```

```python
print(train_X.shape)
print(test_X.shape)
```

```python
lrg = LinearRegression()
lrg_model = lrg.fit(train_X,train_y)
```

```python
print('Intercept Linear Regression:',lrg_model.intercept_)
print("Slope-----------------------------------------------------")
list(zip(X.columns,lrg_model.coef_))
```

```python
pred = lrg_model.predict(test_X)
print(pred)
```

```python
r2_Sc = r2_score(test_y,pred)
print("R2 Score",r2_Sc)
```

```python
rmse = np.sqrt(mean_squared_error(test_y,pred))
print("RMSE value ",rmse)
```

```python
lrg.score(test_X,test_y)
```

The R-squared score of 0.84 indicates that 84 percent of the variance in the test set is explained by the model.

```python
def RMSE_TEST(features_cols):
    X=df[features_cols]
    y=df['RH']
    X_std = StandardScaler().fit_transform(X) 
    train_X,test_X,train_y,test_y = train_test_split(X,y,test_size=0.25,random_state=40)
    lr1=lrg.fit(train_X,train_y)
    y_pred = lr1.predict(test_X)
    return(np.sqrt(mean_squared_error(test_y,y_pred)))
    
```

```python
features_df.remove('RH')
```

```python
print(features_df[0:4],RMSE_TEST(features_df[0:4]))
print(features_df[0:6],RMSE_TEST(features_df[0:6]))
print(features_df[1:4],RMSE_TEST(features_df[1:4]))
print(features_df[0:9],RMSE_TEST(features_df[0:9]))
print(features_df[1:10],RMSE_TEST(features_df[1:10]))
print(features_df[1:13],RMSE_TEST(features_df[1:13]))
print(features_df[1:9],RMSE_TEST(features_df[1:9]))
print(features_df[0:13],RMSE_TEST(features_df[0:13]))
```

# Support Vector Machine Regression

```python
from sklearn.svm import SVR
```

```python
all_kernels=[]
r2_score_svm=[]
def eva_svr(x,y):
    kernels=['linear','poly','rbf','sigmoid']
    for i in kernels:
        sv=SVR(kernel=i)
        sv_model=sv.fit(x,y)
        sv_predict = sv_model.predict(test_X)
        rmse_svm = np.sqrt(mean_squared_error(test_y,sv_predict))
        r2_svm = r2_score(test_y,sv_predict)
        all_kernels.append(rmse_svm)
        r2_score_svm.append(r2_svm)
        
        
```

```python
eva_svr(train_X,train_y)
```
